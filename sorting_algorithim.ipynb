Sorting Algorithim
*   insertion sort
*   selection sort
*   quick sort





```python
#import required modules
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
############## insertion sort ##################
count_ins_swap = 0
def swap(lst,a,b):
	global count_ins_swap
	""" swaps a and b """ 
	lst[a],lst[b] = lst[b],lst[a]
	count_ins_swap +=1

count_ins_com = 0
def push_down(lst,i):                            
	global count_ins_com
	""" Push value at position i into sorted position in b[0..i-1] """
	j = i
	while j > 0 :
		count_ins_com += 1	
		if lst[j-1] > lst[j]:
			swap(lst,j-1,j)

		j = j-1
def ins_sort(lst):
	""" sort the list """
	n = len(lst)
	i = 0
	while i < n:
		push_down(lst,i)
		i += 1
```

```python
lst2  = []
ins_com = []
ins_swap = []

for i in range(40,1000,50):
	lst2.append(i)
	total_swap_ins = 0
	total_com_ins = 0
	for j in range(3):
		lst1 = [random.randint(100,400) for i in range(i)]
		#count_ins_swap = 0
		#count_ins_com = 0
		ins_sort(lst1)
		total_swap_ins += count_ins_swap
		total_com_ins += count_ins_com
		
	ins_swap.append(total_swap_ins/3)
	ins_com.append(total_com_ins/3)
	
print('insertion', ins_swap)
print('insertion',ins_com)

```

```python
# Graph analysis for com and swap of insertion sort
plt.plot(lst2, ins_com, label =  'insertion comparision ')
plt.plot(lst2, ins_swap, label =  'insertion swap ')
plt.legend()
plt.show()
```

```python
################# selection sort ##################
count_sel_com = 0
def find_min(arr,i):
	global count_sel_com
	""" Returns the index of minimum number in the array """
	j = i
	min_element_pos = i 
	while j < len(arr):
		if arr[min_element_pos] > arr[j]:
			count_sel_com += 1
			min_element_pos = j
			j += 1
		j += 1
	return min_element_pos

count_sel_swap = 0	
def selection_sort(lst):
	""" sort the array """
	global count_sel_swap
	i = 0
	j = len(lst)
	while i < j :
		a = find_min(lst,i)
		lst[a], lst[i] = lst[i], lst[a]
		count_sel_swap += 1 
		i += 1
```

```python
lst3 = []
sel_com = []
sel_swap = []

for i in range(40,1000,50):
	lst3.append(i)
	total_swap_sel = 0
	total_com_sel = 0
	for j in range(3):
		lst1 = [random.randint(100,400) for i in range(i)]
		#count_sel_swap = 0
		#count_sel_com = 0
		selection_sort(lst1)
		total_swap_sel += count_sel_swap
		total_com_sel += count_sel_com
		
	sel_swap.append(total_swap_sel/3)
	sel_com.append(total_com_sel/3)
	
print('selection', sel_swap)
print('selection',sel_com)
```

```python
# Graph analysis for com and  swap for selection sort
plt.plot(lst3, sel_com, label =  'selection comparision ')
plt.plot(lst3, sel_swap, label =  'selection swap ')
plt.legend()
plt.show()
```

```python
################# quick sort ##############
count_swap_quick = 0
count_com_quick = 0
def partition(lst,h,k):
	""" Partition list b[h..k] around a pivot x = b[h] """
	global count_com_quick
	global count_swap_quick
	i = h
	j = k
	pivot = lst[h]

	while i < j :
		count_com_quick +=1
		if lst[i+1] <  pivot :
			count_swap_quick += 1
			lst[i],lst[i+1] = lst[i+1],lst[i]
			i += 1

		else:
			count_swap_quick += 1
			lst[i+1],lst[j] = lst[j],lst[i+1]
			j -= 1

	return i




def quick_sort(lst,h,k):
	""" sort the array fragement [h..k] """

	if h > k :
		return

	else:
		p = partition(lst,h,k)
		quick_sort(lst,h,p-1)
		quick_sort(lst,p+1,k)

```

```python
lst4 = []
quick_swap = []
quick_com = []

for i in range(40,1000,50):
	lst4.append(i)
	total_swap_quick = 0
	total_com_quick = 0
	for j in range(3):
		lst1 = [random.randint(100,400) for i in range(i)]
		#count_swap_quick = 0
		#count_com_quick = 0
		quick_sort(lst1,0,len(lst1)-1)
		total_swap_quick += count_swap_quick
		total_com_quick += count_com_quick
		
	quick_swap.append(total_swap_quick/3)
	quick_com.append(total_com_quick/3)
	
print('quick sort', quick_swap)
print('quick sort',quick_com)
```

```python
# Graph analysis for com and swap for quick sort 
plt.plot(lst4, quick_com, label =  'quick comparision ')
plt.plot(lst4, quick_swap, label =  'quick   swap ')
plt.legend()
plt.show()
```

```python
# Combined Graph analysis for swaps of all three sorting algorithm
plt.plot(lst2, ins_swap, label =  'ins swap ')
plt.plot(lst4, quick_swap, label =  'quick   swap ')
plt.plot(lst3, sel_swap, label =  'sel   swap ')
plt.legend()
plt.show()
```
